```python
#preprocessing
normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
train_transformer = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomResizedCrop((224),scale=(0.5,1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    normalize
])

val_transformer = transforms.Compose([
    transforms.Resize(224),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    normalize
])
batchsize=10
def read_txt(txt_path):
    with open(txt_path) as f:
        lines = f.readlines()
    txt_data = [line.strip() for line in lines]
    return txt_data
```

```python
class CovidCTDataset(Dataset):
    def __init__(self, root_dir, txt_COVID, txt_NonCOVID, transform=None):
        
        self.root_dir = root_dir
        self.txt_path = [txt_COVID,txt_NonCOVID]
        self.classes = ['CT_COVID', 'CT_NonCOVID']
        self.num_cls = len(self.classes)
        self.img_list = []
        for c in range(self.num_cls):
            cls_list = [[os.path.join(self.root_dir,self.classes[c],item), c] for item in read_txt(self.txt_path[c])]
            self.img_list += cls_list
        self.transform = transform

    def __len__(self):
        return len(self.img_list)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        img_path = self.img_list[idx][0]
        image = Image.open(img_path).convert('RGB')

        if self.transform:
            image = self.transform(image)
        sample = {'img': image,
                  'label': int(self.img_list[idx][1])}
        return sample



    
if __name__ == '__main__':
    trainset = CovidCTDataset(root_dir='/content/COVID-CT/Images-processed',
                              txt_COVID='/content/COVID-CT/Data-split/COVID/testCT_COVID.txt',
                              txt_NonCOVID='/content/COVID-CT/Data-split/NonCOVID/testCT_NonCOVID.txt',
                              transform= train_transformer)
    valset = CovidCTDataset(root_dir='/content/COVID-CT/Images-processed',
                              txt_COVID='/content/COVID-CT/Data-split/COVID/valCT_COVID.txt',
                              txt_NonCOVID='/content/COVID-CT/Data-split/NonCOVID/valCT_NonCOVID.txt',
                              transform= val_transformer)
    testset = CovidCTDataset(root_dir='/content/COVID-CT/Images-processed',
                              txt_COVID='/content/COVID-CT/Data-split/COVID/testCT_COVID.txt',
                              txt_NonCOVID='/content/COVID-CT/Data-split/NonCOVID/testCT_NonCOVID.txt',
                              transform= val_transformer)
    print(trainset.__len__())
    print(valset.__len__())
    print(testset.__len__())

    train_loader = DataLoader(trainset, batch_size=batchsize, drop_last=False, shuffle=True)
    val_loader = DataLoader(valset, batch_size=batchsize, drop_last=False, shuffle=False)
    test_loader = DataLoader(testset, batch_size=batchsize, drop_last=False, shuffle=False)
```

```python
dataset =ConcatDataset([trainset,testset])
print(len(dataset))
```

```python
#CNN-LSTM Model
class COVID19(nn.Module):
    def __init__(self):
        super().__init__()
        self.network = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, stride=1,padding=1),
            nn.Conv2d(64, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),# output: 112 x 112 x 64
            
            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.Conv2d(128, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), # output: 56 x 56 x 128
            
            nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1),
            nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), #output 28*28*256
            
            nn.Conv2d(256, 512, kernel_size=3, stride=1, padding=1),
            nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), #output 14*14*512

            nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
            nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2, 2), #output 7*7*512
            )
        self.lstm=nn.Sequential(
             nn.LSTM(49,512,1, dropout=0.2, batch_first=True)   
            )
        self.linear=nn.Sequential(
            nn.Flatten(),
            nn.Linear(262144,64),
            nn.ReLU(),
            nn.Linear(64,2),
                
            )
            
            #nn.ReLU(),
            #nn.Linear(512, 10))
            
    def forward(self,x):
        #print(x.shape)
        x = self.network(x)
        #print(x.shape)
        x=x.reshape(10,512,49)
        x,(h,c)=self.lstm(x)
        #print(x.shape)
        x=self.linear(x)
        return x
            

```

```python
model = COVID19()
criterion = CrossEntropyLoss()



if torch.cuda.is_available():
    model = model.cuda()
    criterion = criterion.cuda()
    
print(model)
```

```python
#Training Model
@torch.no_grad()
def evaluate(model, val_loader):
    model.eval()
    outputs = [model.validation_step(batch) for batch in val_loader]
    return model.validation_epoch_end(outputs)
def fit(epochs, model, train_loader):
    bs=10
    right = []
    criterion = CrossEntropyLoss()
    history = []
    optimizer = Adam(model.parameters(), lr=0.0001)
    train_loss_min=np.Inf
    for epoch in range(epochs):
      model.train()
      train_loss = 0
      train_correct = 0
      val_correct = 0
      val_loss=0
      for idx,batch in enumerate(train_loader):
        if batch['label'].shape[0]%10!=0:
          continue
        x_train,y_train= batch['img'].cuda(),batch['label'].cuda()
        outputs=model(x_train)
        loss=criterion(outputs,y_train)
        train_loss += criterion(outputs,y_train.long())
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(),0.2)
        optimizer.step()
        optimizer.zero_grad()
            
      train_loss1=train_loss/(len(train_loader.dataset)-3)
      print('Epoch: {} \tTraining Loss: {:.6f}'.format(epoch, train_loss1))
      if train_loss1 <= train_loss_min:
              print('Train loss decreased ({:.6f} --> {:.6f}).  Saving model ...'.format(train_loss_min,train_loss1))
              torch.save(model.state_dict(), 'model_covidct.pt')#Saving Model With least loss
              train_loss_min = train_loss1
```
